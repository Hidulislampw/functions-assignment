1. What is the difference between a function and a method in Python?
Ans 1.  In Python, a function is a reusable block of code defined using the def keyword that operates independently and can be called anywhere in the program. A method, on the other hand, is a function that is associated with an object and is called on that object. Methods are typically defined inside classes and operate on the instance (or class) they belong to, often modifying or interacting with the object's attributes.

```python
#For example:

def greet(name):
    return f"Hello, {name}!"

print(greet("Hidul"))  # Calling a function

# Method inside a class
class Person:
    def __init__(self, name):
        self.name = name

    def greet(self):
        return f"Hello, {self.name}!"

p = Person("Bob")
print(p.greet())  # Calling a method on an object
```

2. Explain the concept of function arguments and parameters in Python.
Ans 2. In Python, parameters are the variables listed in a function's definition, while arguments are the actual values passed to the function when it is called. Python supports different types of function arguments: positional arguments (passed in order), keyword arguments (explicitly named), default arguments (having predefined values), and arbitrary arguments (*args for multiple positional arguments and **kwargs for multiple keyword arguments). This flexibility allows functions to be highly adaptable to different use cases.

```python
def introduce(name, age=25, *hobbies, **details):
    print(f"Name: {name}")
    print(f"Age: {age}")
    print(f"Hobbies: {', '.join(hobbies)}")
    print(f"Other details: {details}")

# Calling the function with different argument types
introduce("Alice", 30, "reading", "traveling", city="New York", profession="Engineer")
```

3.  What are the different ways to define and call a function in Python?
Ans 3. In Python, functions can be defined in multiple ways, including regular functions, lambda functions, and functions inside classes (methods). A regular function is defined using the def keyword, followed by a function name and optional parameters. It can be called by passing the required arguments. Lambda functions, also known as anonymous functions, are defined using the lambda keyword and can be used for short, one-line operations. Additionally, functions can be defined within classes as instance or class methods and called using an instance of the class. Python also supports higher-order functions, where functions are passed as arguments, and nested functions, where functions are defined inside other functions.

```python
# 1. Regular function
def add(a, b):
    return a + b

print(add(5, 3))  # Calling the function

# 2. Lambda function
multiply = lambda x, y: x * y
print(multiply(4, 6))  # Calling the lambda function

# 3. Function inside a class (Method)
class Calculator:
    def subtract(self, a, b):
        return a - b

calc = Calculator()
print(calc.subtract(10, 4))  # Calling the method

# 4. Higher-order function (passing function as argument)
def operate(func, x, y):
    return func(x, y)

print(operate(add, 7, 2))  # Calling a function using another function

# 5. Nested function
def outer_function(msg):
    def inner_function():
        return f"Inner says: {msg}"
    return inner_function()

print(outer_function("Hello!"))  # Calling a nested function
```

5.  What are iterators in Python and how do they differ from iterables?
Ans 5. In Python, an iterable is any object that can return an iterator, meaning it can be looped over using a for loop. Examples include lists, tuples, strings, and dictionaries, which implement the __iter__() method. An iterator, on the other hand, is an object that keeps track of its position and returns elements one at a time using the __next__() method. Iterators are created by calling iter() on an iterable, and they remember their current state, making them useful for working with large datasets efficiently. Unlike iterables, iterators do not restart once exhausted.

```python
# Iterable: A list
numbers = [1, 2, 3, 4]
for num in numbers:  # Iterating over an iterable
    print(num)

# Iterator: Using iter() and next()
numbers_iter = iter(numbers)  # Creating an iterator from a list
print(next(numbers_iter))  # Output: 1
print(next(numbers_iter))  # Output: 2
```

6. Explain the concept of generators in Python and how they are defined.
Ans 6. In Python, a generator is a special type of iterable that allows lazy evaluation, meaning it generates values one at a time as they are requested instead of storing them in memory. Generators are defined using functions with the yield keyword instead of return. When a generator function is called, it does not execute immediately but returns a generator object, which can be iterated using a loop or the next() function. Each time yield is encountered, the function's state is saved, and execution resumes from that point when the generator is called again. This makes generators memory-efficient and useful for handling large data sets or infinite sequences.

```python
# Generator function to yield squares of numbers
def square_numbers(n):
    for i in range(n):
        yield i * i  # Yielding square of i

# Using the generator
gen = square_numbers(5)
print(next(gen))  # Output: 0
print(next(gen))  # Output: 1
print(next(gen))  # Output: 4

# Iterating using a loop
for num in square_numbers(5):
    print(num)  # Outputs: 0, 1, 4, 9, 16
```

7. What are the advantages of using generators over regular functions?
Ans 7. Generators in Python provide an efficient way to handle large datasets and iterative computations without consuming excessive memory. Unlike regular functions that return a value and terminate, generators use the yield keyword to return a sequence of values lazily, meaning they generate values on demand instead of storing them all in memory at once. This makes generators particularly useful for handling large data streams, infinite sequences, and cases where the entire dataset doesnâ€™t need to be loaded at once. Generators also improve performance and readability by maintaining state between iterations without needing explicit memory management.

```python
# Regular function (returns a list, consuming more memory)
def square_numbers(n):
    result = []
    for i in range(n):
        result.append(i ** 2)
    return result

print(square_numbers(5))  # [0, 1, 4, 9, 16]

# Generator function (yields values one by one)
def square_numbers_gen(n):
    for i in range(n):
        yield i ** 2  # Generates values lazily

gen = square_numbers_gen(5)
print(next(gen))  # 0
print(next(gen))  # 1
print(next(gen))  # 4
```

8. What is a lambda function in Python and when is it typically used?
Ans 8. A lambda function in Python is an anonymous, single-expression function defined using the lambda keyword. Unlike regular functions created with def, lambda functions do not have a name and are typically used for short, simple operations where defining a full function is unnecessary. They are often used in scenarios requiring a small function for immediate use, such as within map(), filter(), or sorted() functions. Since they have an implicit return statement, they are useful for quick, inline computations.

```python
# Regular function
def square(x):
    return x * x

print(square(5))  # Output: 25

# Lambda function
square_lambda = lambda x: x * x
print(square_lambda(5))  # Output: 25

# Using lambda inside map()
numbers = [1, 2, 3, 4]
squared_numbers = list(map(lambda x: x * x, numbers))
print(squared_numbers)  # Output: [1, 4, 9, 16]
```

9. Explain the purpose and usage of the `map()` function in Python.
Ans 9. The map() function in Python is used to apply a given function to all items in an iterable (such as a list or tuple) and return a map object (an iterator). It is particularly useful for performing transformations on data without using explicit loops, making the code more concise and efficient. The syntax is map(function, iterable), where function is applied to each element of iterable. If multiple iterables are passed, map() applies the function by taking corresponding elements from each iterable. The result can be converted into a list, tuple, or other iterable types.

```python
# Using map() to square numbers in a list
def square(num):
    return num ** 2

numbers = [1, 2, 3, 4, 5]
squared_numbers = list(map(square, numbers))  # Convert map object to list
print(squared_numbers)  # Output: [1, 4, 9, 16, 25]

# Using map() with lambda function
doubled_numbers = list(map(lambda x: x * 2, numbers))
print(doubled_numbers)  # Output: [2, 4, 6, 8, 10]
```

10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
Ans 10. The map(), reduce(), and filter() functions in Python are built-in higher-order functions that operate on iterables, applying a function to each element. The map() function applies a given function to all items in an iterable and returns a new iterable with the transformed values. The filter() function filters elements from an iterable based on a condition defined in the given function, returning only the elements that evaluate to True. The reduce() function (from the functools module) applies a function cumulatively to the elements of an iterable, reducing it to a single value.

```python
from functools import reduce

# map() example - squares each number
numbers = [1, 2, 3, 4, 5]
squared = list(map(lambda x: x ** 2, numbers))
print(squared)  # Output: [1, 4, 9, 16, 25]

# filter() example - filters even numbers
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
print(even_numbers)  # Output: [2, 4]

# reduce() example - calculates the product of all numbers
product = reduce(lambda x, y: x * y, numbers)
print(product)  # Output: 120 (1*2*3*4*5)
```

```python
#4. Write a Python function that checks if a given number is prime or not from 1 to 200.
def is_prime(n):
    if n < 2:
        return False  # 0 and 1 are not prime numbers
    for i in range(2, int(n ** 0.5) + 1):  # Check divisibility up to sqrt(n)
        if n % i == 0:
            return False  # If divisible, it's not prime
    return True  # If no divisors found, it's prime

# Example usage:
for num in range(1, 201):
    if is_prime(num):
        print(num, end=" ")
```

```python
#5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
class FibonacciIterator:
    def __init__(self, n_terms):
        self.n_terms = n_terms  # Total terms to generate
        self.a, self.b = 0, 1  # Initial two Fibonacci numbers
        self.count = 0  # Counter to track iterations

    def __iter__(self):
        return self  # The class itself is the iterator

    def __next__(self):
        if self.count >= self.n_terms:  # Stop iteration when limit is reached
            raise StopIteration
        fib_number = self.a  # Store the current Fibonacci number
        self.a, self.b = self.b, self.a + self.b  # Move to the next numbers
        self.count += 1  # Increment counter
        return fib_number  # Return the Fibonacci number

# Example usage:
fib_iter = FibonacciIterator(10)  # Create an iterator for 10 Fibonacci numbers
for num in fib_iter:
    print(num, end=" ")
```

```python
#6.  Write a generator function in Python that yields the powers of 2 up to a given exponent.
def powers_of_two(max_exponent):
    for exponent in range(max_exponent + 1):  # Iterate from 0 to max_exponent
        yield 2 ** exponent  # Yield 2 raised to the power of exponent

# Example usage:
for power in powers_of_two(5):
    print(power, end=" ")

```

```python
#7.  Implement a generator function that reads a file line by line and yields each line as a string.
def read_file_line_by_line(filename):
    try:
        with open(filename, 'r', encoding='utf-8') as file:
            for line in file:
                yield line.strip()  # Yield each line after stripping leading/trailing spaces
    except FileNotFoundError:
        print(f"Error: The file '{filename}' was not found.")
    except Exception as e:
        print(f"Error: {e}")

# Example usage:
filename = "sample.txt"  # Replace with your file name
for line in read_file_line_by_line(filename):
    print(line)
```
